```python
import pandas as pd
import numpy as np
```

```python
from nba_api.stats.endpoints import drafthistory
```

```python
players = pd.DataFrame()

for season_year in range(1980, 2016):
    for overall_pick in range(1,11):
        player = drafthistory.DraftHistory(overall_pick_nullable=overall_pick, season_year_nullable=season_year).get_data_frames()[0][['PERSON_ID', 'PLAYER_NAME', 'SEASON', 'OVERALL_PICK']]
        players = pd.concat([players, player], ignore_index=False)

players
    
```

```python
players.to_csv('players_1980-2015')
```
