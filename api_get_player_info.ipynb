```python
import pandas as pd
from nba_api.stats.endpoints import commonplayerinfo
```

```python
nba_players = pd.read_csv("nba_players")
```

```python
player_ids = nba_players['PERSON_ID']
```

```python
player_info = pd.DataFrame()

for player_id in player_ids:
    info = commonplayerinfo.CommonPlayerInfo(player_id=player_id)
    df_info = info.get_data_frames()[0]
    player_info = pd.concat([player_info, df_info], ignore_index=False)

# Display player details
player_info
```

```python
player_info.to_csv("player_info.csv", index=False)
```
