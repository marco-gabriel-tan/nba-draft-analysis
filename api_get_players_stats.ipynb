```python
import pandas as pd
from nba_api.stats.endpoints import playercareerstats
```

```python
nba_players = pd.read_csv('nba_players')
nba_players = nba_players.drop(columns=['Unnamed: 0'])
nba_players
```

```python
nba_players[nba_players['SEASON'] == 1978]
```

```python
player_ids = nba_players['PERSON_ID']
player_ids
```

```python
player_stats = pd.DataFrame()
for player_id in enumerate(player_ids):
    player = playercareerstats.PlayerCareerStats(player_id=player_id[1]).get_data_frames()[0]
    player_stats = pd.concat([player_stats, player], ignore_index=False)

player_stats
```

```python
player_stats.to_csv('player_stats.csv', index=False)
```
