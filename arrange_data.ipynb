```python
#### Adds Index Column and Points per game, Rebounds per game, Assists per game,
#### Steals per game, Blocks per game, and Turnovers per game.

import pandas as pd
```

```python
nba_stats = "player_stats.csv"
```

```python
player_stats = pd.read_csv(nba_stats)
```

```python
player_stats
```

```python
player_stats['PPG'] = player_stats.apply(lambda x: round(x['PTS'] / x['GP'],2), axis=1)
player_stats['RPG'] = player_stats.apply(lambda x: round(x['REB'] / x['GP'], 2), axis=1)
player_stats['APG'] = player_stats.apply(lambda x: round(x['AST'] / x['GP'], 2), axis=1)
player_stats['SPG'] = player_stats.apply(lambda x: round(x['STL'] / x['GP'], 2), axis=1)
player_stats['BPG'] = player_stats.apply(lambda x: round(x['BLK'] / x['GP'], 2), axis=1)
player_stats['TPG'] = player_stats.apply(lambda x: round(x['TOV'] / x['GP'], 2), axis=1)
```

```python
player_stats
```

```python
player_stats.to_csv("arranged_player_stats.csv", index=False)
```
